```python
import requests
import json 
import pandas as pd


URL="http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes"


req2 = requests.get(URL)
text_data2= req2.text
json_dict2= json.loads(text_data2)



df= pd.DataFrame.from_dict(json_dict2['_embedded']['episodes']).drop(labels=["summary","_links"],axis=1)

#df2=df2.drop(labels=["summary","_links"],axis=1)
```

```python
df["runtime"]=df["runtime"].astype(float)
```

```python
df["airdate"]=pd.to_datetime(df["airdate"],format="%Y-%m-%d")
```
